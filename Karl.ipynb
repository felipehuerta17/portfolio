# 7. import functions from another library

```python
# Get bokeh function to plot a figure and to read data arranged in columns
from bokeh.plotting import figure, ColumnDataSource
# Get bokeh function for outputs that we want
from bokeh.io import output_file, show, output_notebook
# Get bokeh function to use HoverTool
from bokeh.models import HoverTool

# Display all bokeh figues inline (inside jupyter notebook)
output_notebook()
```

Load the dataframe data to be used within the bokeh library

```python
source = ColumnDataSource(data)
```

Create bokeh fugure with the HoverTool that is empty (for now)

```python
p=figure(tools=['hover'])
```

Create the figure p  - displaying the datapoints as dots (ie circles)

```python
p.circle(x='market_value', y='page_views', source=source)
# specify the hovertool
hover=HoverTool(tooltips=[('Name', '@name'),('Team', '@club')])
```

```python
p.add_tools(hover)
show(p)
```
